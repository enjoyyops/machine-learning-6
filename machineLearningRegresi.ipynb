```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Dataset
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Model
model = LinearRegression()
model.fit(X, y)

# Prediksi
y_pred = model.predict(X)

# Parameter model
print("Koefisien (slope):", model.coef_[0][0])
print("Intersep:", model.intercept_[0])
```

```python
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
mape = np.mean(np.abs((y - y_pred) / (y + 1e-10))) * 100

print("\n=== EVALUASI REGRESI ===")
print("R2 Score :", r2)
print("MSE      :", mse)
print("RMSE     :", rmse)
print("MAE      :", mae)
print("MAPE     :", mape)

# Plot (dengan garis regresi merah)
plt.scatter(X, y, label="Data Asli")
plt.plot(X, y_pred, color='red', linewidth=2, label="Garis Regresi")
plt.title("Regresi Linear")
plt.xlabel("Fitur")
plt.ylabel("Target")
plt.legend()
plt.show()
```
